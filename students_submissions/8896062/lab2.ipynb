**SCATTER PLOT**
This is scatter plot using data set iris.

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("iris")
sns.scatterplot(x="sepal_length", y="sepal_width", hue="species", data=data)
plt.title("Scatter Plot Example")
plt.show()
```

**Line plot**
This is line plot using variables x and y.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(x, y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot Example")
plt.show()

```

**Line plot**
This is line plot using variables x and y.

```python
import plotly.graph_objects as go

fig = go.Figure(data=go.Scatter(x=[1, 2, 3], y=[4, 5, 6]))
fig.update_layout(title='Line Plot Example', xaxis_title='X-axis', yaxis_title='Y-axis')
fig.show()
```

**SCATTER PLOT**
This is scatter plot using variables x and y.

```python
import plotly.express as px

df = px.data.iris()
fig = px.scatter(df, x="sepal_width", y="sepal_length", color="species",
                 size='petal_length', hover_data=['petal_width'])
fig.show()

```
