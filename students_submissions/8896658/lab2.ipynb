 **LINE GRAPH**

 This line plot, created using Matplotlib, visualizes the trend of a variable across a continuous range.
 created a subplot with two plots using plt.subplots(2, 1), which creates a 2x1 grid of subplots. 

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y1 = [3, 2, 7, 1, 10]
y2 = [1, 3, 5, 7, 9]

# Create a subplot with 1 row and 2 columns
fig, (ax1, ax2) = plt.subplots(1, 2)

# Plot the first data on the first subplot
ax1.plot(x, y1, color="r")
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis')
ax1.set_title('Figure 1')

# Plot the second data on the second subplot
ax2.plot(x, y2,color="g")
ax2.set_xlabel('X-axis')
ax2.set_ylabel('Y-axis')
ax2.set_title('Figure 2')

# Adjust the spacing between subplots
plt.tight_layout()

# Display the plot
plt.show()
```

**Scatter Plot**

A scatter plot represents the relationship between two numerical variables. This example uses Seaborn.
Seaborn provides easy access to a few datasets that are bundled with the library itself.
These functions allow you to load the corresponding dataset into a Pandas DataFrame, which can then be used for data analysis and visualization using Seaborn.

```python
import seaborn as sns
sns.get_dataset_names()
```

```python
tips=sns.load_dataset("tips")
tips
```

```python
sns.scatterplot(x="tip",y="total_bill",data=tips,hue='day',size="size",palette="YlGnBu")
```

```python
sns.stripplot(x="day",y="tip",data=tips,hue="sex",dodge=True)
```

**BAR CHART**

A bar chart is a great way to compare categorical data. This example uses Matplotlib.The function creates a bar plot bounded with a rectangle depending on the given parameters

```python

import numpy as np
import matplotlib.pyplot as plt
 
  
# creating the dataset
data = {'C':20, 'C++':10, 'Java':30,
        'Python':35}
courses = list(data.keys())
values = list(data.values())
  
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(courses, values, color ='green',width = 0.3)
 
plt.xlabel("Courses offered")
plt.ylabel("No. of students enrolled")
plt.title("Students enrolled in different courses")
plt.show()
```

**Pie Chart**

Pie charts are useful for illustrating proportions or percentages. This example uses Plotly.

```python
import plotly.graph_objects as go
import plotly
plotly.offline.init_notebook_mode()
labels = ['Category A', 'Category B', 'Category C']
values = [30, 40, 30]

fig = go.Figure(data=[go.Pie(labels=labels, values=values)])
fig.update_layout(title='Pie Chart with Legend')
fig.show()
```
