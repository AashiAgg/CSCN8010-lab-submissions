**Frame the problem and look at the big picture.**

Problem Framing:

The problem at hand is to perform a univariate linear regression using the diabetes dataset, specifically focusing on the relationship between BMI (body mass index) and a quantitative measure of disease progression one year after baseline. The goal is to build a model that can predict disease progression based on BMI.

big picture:

Understanding this relationship can be valuable for doctors, scientists, and policymakers. It can help us learn how being overweight or obese (as indicated by a higher BMI) influences the advancement of diabetes. This knowledge can improve how healthcare providers manage and treat the disease in their patients. It may also lead to the development of targeted interventions and strategies to combat the negative effects of obesity on diabetes progression.

To investigate the relationship between BMI and disease progression, we will use a method called univariate linear regression. This technique allows us to examine the connection between one independent variable (BMI) and one dependent variable (disease progression) in a straightforward way.

Our analysis will involve examining the parameters of the linear regression model, such as the slope and intercept, to understand how BMI influences disease progression. We will also evaluate the model's performance by measuring the loss, which represents the difference between the predicted and actual values of disease progression, on both the training and test datasets.

However, it's important to note that a linear regression model may not capture all the complexities of the relationship between BMI and diabetes progression. Other factors, like genetics, lifestyle, and medical history, could also play a role. Therefore, we may need to consider additional variables and explore more advanced modeling techniques to gain a deeper understanding of diabetes progression.

In summary, our aim is to use univariate linear regression to explore the connection between BMI and diabetes progression. This analysis can contribute to our broader understanding of diabetes and help healthcare providers make informed decisions about patient care.

**Get the data.**

```python
# Import necessary libraries
from sklearn import datasets
from sklearn.model_selection import train_test_split 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)
print(diabetes_X)
print("---------------")
print(diabetes_y)
```

```python
x = diabetes_X[:, 2] # Use BMI as the input variable
y = diabetes_y
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x,y ,test_size=0.2, random_state=0)
```

**Explore the data to gain insights.**

```python
# Plot the scatter plot to visualize the relationship between BMI and the target variable
plt.scatter(x, y ,color='r')
plt.xlabel('BMI')
plt.ylabel('Disease Progression')
plt.title('Diabetes Dataset - BMI vs Disease Progression')
plt.show()
```

```python
# Create a histogram of the disease progression variable
plt.hist(y, bins=20 ,color='teal')
plt.xlabel('BMI')
plt.ylabel('Disease Progression')
plt.title('Diabetes Dataset - BMI vs Disease Progression')
plt.show()
```

```python
# Create a box plot of  Disease Progression
plt.boxplot(y)
plt.xlabel('BMI')
plt.ylabel('Disease Progression')
plt.title('Diabetes Dataset - BMI vs Disease Progression')
plt.show()
```

```python
# Calculate the correlation coefficient between BMI and the target variable
correlation = np.corrcoef(x, y)[0, 1]
print("Correlation coefficient:", correlation)
```

**Fine-tune your models and combine them into a great solution.**

```python
# Create a linear regression model

model = LinearRegression()

# Fit the model on the training data
model.fit(X_train, y_train)

# Predict on training and test data
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# Calculate mean squared error (loss) on training and test data

train_loss = mean_squared_error(y_train, y_train_pred)
test_loss = mean_squared_error(y_test, y_test_pred)
```

```python
# Plot the training data, test data, and the model
plt.scatter(X_train, y_train, color='y', label='Training Data')
plt.scatter(X_test, y_test, color='red', label='Test Data')
plt.plot(X_train, y_train_pred, color='black', linewidth=2, label='Linear Regression')
plt.xlabel('BMI')
plt.ylabel('Disease Progression')
plt.legend()
plt.show()
```

```python
# Print the results
print("Model Parameters:")
print("Slope: {:.2f}".format(model.coef_[0]))
print("Intercept: {:.2f}".format(model.intercept_))
print("Train Loss: {:.2f}".format(train_loss))
print("Test Loss: {:.2f}".format(test_loss))
```
