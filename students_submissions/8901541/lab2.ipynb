#**Bar Plot of Air Quality Measurements**<br>
->Bar plot using Matplotlib to represent air quality measurements of different pollutants.<br> ->The sample data includes pollutant categories (CO, NO2, SO2, O3) and their corresponding measurement values.<br> ->The bar plot visually displays the measurements for each pollutant category using bars of equal width.<br> ->Axes labels and a title are added to provide context to the plot.

```python

import numpy as np
import matplotlib.pyplot as plt

# Sample data
categories = ['CO', 'NO2', 'SO2', 'O3']
values = [0.5, 0.8, 0.3, 0.6]

# Create bar plot
plt.bar(categories, values, color='blue')

# Set labels and title
plt.xlabel('Pollutant')
plt.ylabel('Measurement')
plt.title('Air Quality Measurements')

# Show the plot
plt.show()
```

#**3D surface plot using Matplotlib**<br>
->This code generates a 3D surface plot using matplotlib. <br>->It creates a surface plot of a mathematical function in three dimensions.<br> ->The plot is created using sample data generated with NumPy and displayed using the plot_surface function.<br> ->The plot is labeled with x, y, and z axes and titled as "3D Surface Plot".<br> ->The resulting plot provides a visual representation of the surface defined by the mathematical function.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate sample data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create a 3D figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='viridis')

# Add labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Surface Plot')

# Show the plot
plt.show()
```

#**1. Matplotlib - 3D Scatter Plot:**<br>
->The Matplotlib code generates a 3D scatter plot of the air quality measurements.<br> ->It uses randomly generated sample data for the x, y, and z coordinates. <br>->Each point in the scatter plot represents a specific measurement, with the color representing the data value. <br>->This plot provides a visual understanding of the distribution of measurements in 3D space.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate sample data
np.random.seed(0)
n = 100
x = np.random.randn(n)
y = np.random.randn(n)
z = np.random.randn(n)

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c='b', marker='o')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title('Air Quality Measurements - 3D Scatter Plot')
plt.show()
```

#**Seaborn - 3D Surface Plot:**<br>
->The Seaborn code creates a 3D surface plot based on the air quality measurements. <br>->It uses sample data for the x, y, and z coordinates, where z values are computed using a mathematical function. <br>->The plot visualizes the surface formed by connecting the data points, allowing us to analyze patterns and variations in the measurements.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate sample data
np.random.seed(0)
n = 50
x = np.linspace(-5, 5, n)
y = np.linspace(-5, 5, n)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create a 3D surface plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title('Air Quality Measurements - 3D Surface Plot')
plt.show()
```

#**Plotly - 3D Line Plot:**<br>
->The Plotly code generates a 3D line plot representing the air quality measurements.<br> ->It uses sample data for the x, y, and z coordinates, where the z values are randomly generated.<br> ->The line plot connects the points in 3D space, allowing us to observe trends and changes in the measurements over the given range.

```python


import plotly.graph_objects as go
import numpy as np

# Generate sample data
np.random.seed(0)
n = 100
x = np.linspace(0, 10, n)
y = np.random.randn(n)
z = np.random.randn(n)

# Create a 3D line plot using Plotly
fig = go.Figure(data=[go.Scatter3d(x=x, y=y, z=z, mode='lines')])
fig.update_layout(scene=dict(xaxis_title='X', yaxis_title='Y', zaxis_title='Z'),
                  title='Air Quality Measurements - 3D Line Plot')
fig.show()
```
