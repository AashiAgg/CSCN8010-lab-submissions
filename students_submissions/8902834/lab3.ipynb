```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)
```

```python

diabetes_bmi = diabetes_X[:,2].reshape((-1,1))
X_train,X_test,y_train,y_test = train_test_split(diabetes_bmi, diabetes_y, test_size=0.2, random_state= 0)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
m = model.coef_
c = model.intercept_
y_pred = model.predict(X_test)
print (m)
print(c)
```

```python
# Plot the training data and the model
plt.scatter(X_train, y_train, color='magenta', label='Training Data')
plt.xlabel('BMI')
plt.ylabel('Disease Progression')
plt.title('Linear Regression - Training Data')
plt.legend()
plt.show()

```

```python
# Plot the training data and the model
plt.scatter(X_test, y_test, color='blue', label='Testing Data')
plt.xlabel('BMI')
plt.ylabel('Disease Progression')
plt.title('Linear Regression - Testing Data')
plt.legend()
plt.show()
```

```python
# Plot the test data and the model
plt.plot(X_test, y_pred, color='red', linewidth=2, label='Linear Regression Model')
plt.xlabel('BMI')
plt.ylabel('Disease Progression')
plt.title('Linear Regression - Test Data')
plt.legend()
plt.show()
```

```python
# Calculate the mean squared error (loss) on the test data
test_loss = np.mean((y_test - y_pred) ** 2)
print("Test Loss:", test_loss)

# Plot the test data and the linear regression model
plt.scatter(X_train, y_train, color='magenta', label='Training Data')
plt.scatter(X_test, y_test, color='blue', label='Test Data')
plt.plot(X_test, y_pred, color='red', linewidth=2, label='Linear Regression Model')
plt.xlabel('BMI')
plt.ylabel('Disease Progression')
plt.title('Linear Regression - Test Data')
plt.legend()
plt.show()
```
