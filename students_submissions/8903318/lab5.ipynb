```python
from sklearn.datasets import load_diabetes
from sklearn.metrics import r2_score, mean_absolute_error
from sklearn.model_selection import cross_val_score
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression
import numpy as np
```

#### Loading the datasets

```python
# load diabetes dataset
Xdata, ydata = load_diabetes(as_frame=True, scaled=False, return_X_y=True)
Xdata
```

### 1. Utilize the diabetes dataset from lab 4. Perform cross-validation on nine polynomial models, ranging from degree 0 to 8.

```python
# define polynomial model
def cross_val_polynomial(degree):
    model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
    scores = cross_val_score(model, Xdata, ydata, cv=5, scoring='neg_mean_squared_error')
    mse = -np.mean(scores)
    model.fit(Xdata, ydata)
    y_pred = model.predict(Xdata)
    r2 = r2_score(ydata, y_pred)
    mae = mean_absolute_error(ydata, y_pred)
    return mse, r2, mae
```

```python
# create polynomial models ranging degree from 0 to 8
results = []
degrees = list(range(9))
cv_mse = []
cv_r2 = []
cv_mae = []

for degree in degrees:
    mse, r2, mae = cross_val_polynomial(degree)
    results.append((degree, mse, r2, mae))
    cv_mse.append(mse)
    cv_r2.append(r2)
    cv_mae.append(mae)
```

### 2. Construct a table summarizing the cross-validation results. Each model should have a separate row in the table. Include the R-Squared and Mean Absolute Error (MAE) metrics for each model. Calculate the mean value and standard deviation of these metrics from the cross-validation. Include both values.

```python

#cross validation results
df_results = pd.DataFrame(
    results, columns=['Degree', 'MSE', 'R-Squared', 'MAE'])

df_results
```

```python
# Summary
mean_mse = np.mean(cv_mse)
mean_r2 = np.mean(cv_r2)
mean_mae = np.mean(cv_mae)
std_mse = np.std(cv_mse)
std_r2 = np.std(cv_r2)
std_mae = np.std(cv_mae)

summary = pd.DataFrame(
    {'Metric': ['MSE', 'R-Squared', 'MAE'],
     'Mean': [mean_mse, mean_r2, mean_mae],
     'Standard Deviation': [std_mse, std_r2, std_mae]
     }
)

summary
```

### 3. Identification of the Best Model: Identify the model that exhibits the highest performance based on the R-Squared and MAE metrics. Provide an explanation for choosing this specific model.

```python
best_model_r2 = df_results['R-Squared'].idxmax()
best_model_mae = df_results['MAE'].idxmin()
print(f"\nBest model based on R-Squared: Degree {best_model_r2}")
print(f"Best model based on MAE: Degree {best_model_mae}")
```
